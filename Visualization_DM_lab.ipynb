## Line Chart 

```python
x = np.linspace(0, 5, 11)
y = x ** 2  
z= x*x
```

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('My first Plot')
plt.show()
```

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('My first Plot')
plt.grid()
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') 
plt.subplot(1,2,2)
plt.plot(z, x, 'g*-');
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,3,1)
plt.plot(x, y, 'r--') 
plt.subplot(1,3,2)
plt.plot(z, x, 'g*-')
plt.subplot(1,3,3)
plt.plot(z, x, 'y*-');
```

```python
days = [1,2,3,4,5]
Enfield = [50,40,70,80,20]
Honda = [80,20,20,50,60]
Yahama = [70,20,60,40,60]
KTM = [80,20,20,50,60]


plt.plot(days,Enfield,'g',label='Enfield')
plt.plot(days,Honda,'y',label='Honda')
plt.plot(days,Yahama ,'b',label='Yahama')
plt.plot(days,KTM,'r',label='KTM')

plt.title('bike details in line plot')
plt.ylabel('Distance in kms')
plt.xlabel('Days')
plt.legend()

```

# Bar Chart 

```python
country = ['A', 'B', 'C', 'D', 'E']
gdp_per_capita = [45000, 42000, 52000, 49000, 47000]

plt.bar(country, gdp_per_capita)
plt.title('Country Vs GDP Per Capita')
plt.xlabel('Country')
plt.ylabel('GDP Per Capita')
plt.show()
```

# Pie Chart 

```python
data = {'Python':29,
        'Javascript':26,
        'java':21,
        'C':17,
        'C++':11}

fig = plt.figure(figsize =(10, 7))
plt.pie(list(data.values()),labels=data.keys())
plt.show()


```

```python
highlight =(0,0.2,0,0,0)
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(list(data.values()),labels=data.keys(),explode=highlight)
plt.show()

```
