Pandas treat None and NaN as essentially interchangeable for indicating missing or null values. There are several useful functions for detecting, removing, and replacing null values in Pandas DataFrame :

    isnull()
    notnull()
    dropna()
    fillna()

```python
# dictionary of lists

dict1 = {'First Score':[100, 90, np.nan, 95,66,55],
        'Second Score': [30, 45, 56, np.nan,44,np.nan],
        'Third Score':[44, 40, 80, 98,17,56],
        'Fourth Score':[np.nan, 40, 80, 98,np.nan,np.nan],
        'Five Score':[np.nan, 40, 80, 48,np.nan,np.nan],
        'Six Score':[np.nan, 60, 80,np.nan,60,69]
        }
```

```python
dict1
```

```python
len(dict1)
```

```python
# filling missing value using fillna() 
df2.fillna(0)
```

```python
# Importing the SimpleImputer class
from sklearn.impute import SimpleImputer

# Imputer object using the mean strategy and
# missing_values type for imputation

imputer = SimpleImputer(missing_values = np.nan,strategy ='mean')
df_111.Age= imputer.fit_transform(df_111[['Age']])


print(" Missing values changed ")
```
